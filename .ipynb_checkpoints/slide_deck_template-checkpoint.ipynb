## Investigation Overview

> In this investigation , I will demonstrate the relationship between the Prosper Rating and some of the other aspects in this data set.

## Dataset Overview

> This data set contains 113,937 loans with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others. [Loan Data Last updated 03/11/2014] .

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('prosperLoanData.csv')
#select a fewer columns to choose the features of interest

df2 = df[['ListingKey','ProsperRating (Alpha)', 'EmploymentStatus', 'BorrowerAPR', 
                  'StatedMonthlyIncome', 'Term','LoanOriginalAmount', 'DebtToIncomeRatio', 'TotalTrades', 
                  'IncomeRange' , 'IsBorrowerHomeowner' , 'CurrentCreditLines' , 'OpenRevolvingAccounts']]

#remove the duplicates
df2.drop_duplicates(subset=['ListingKey'], inplace=True)

#drop the missing rows 
df2 = df2[~ df2['ProsperRating (Alpha)'].isna()]
df2 = df2[~ df2['DebtToIncomeRatio'].isna()]
df2 = df2[~ df2['EmploymentStatus'].isna()]
```

## The number of persons who choose each prosper rating type

> Several people are taking a medium amount of risk when making investments or taking out loans.

```python
order = ['AA', 'A', 'B', 'C', 'D', 'E', 'HR']
plt.figure(figsize=(8, 6))
color_base = sb.color_palette()[0]
sb.countplot( data = df2 , x ='ProsperRating (Alpha)' , order = order , color = color_base);
plt.title("The number of ProsperRating (Alpha) for each type ");
```

## the number of people with each Income Range across all Prosper Rating types

> People with a salary of 50,000 to 75,000 likely to take modest to moderately risks. People with a salary of 25,000 to 50,000, on the other hand, tend to take significantly greater risks.

```python
plt.figure(figsize=(15,5))
income_order = ['$100,000+', '$75,000-99,999', '$50,000-74,999','$25,000-49,999', '$1-24,999']
sb.countplot(data = df2, x = 'ProsperRating (Alpha)', 
              hue = 'IncomeRange', order=order, hue_order=income_order);
plt.title('add up the Income Ranges for each Prosper Rating (Alpha)');
```

## the proportion of persons who own a home or do not own a home across all Prosper Rating types

> Surprisingly, owning a property has no influence on risk rating!

```python
plt.figure(figsize = [15,5])
sb.countplot(data = df2, x = 'ProsperRating (Alpha)', hue = 'IsBorrowerHomeowner', order=order );
plt.title('Count If Borrower Homeowner for each Prosper Rating (Alpha)');
```

## Borrower APR for all Prosper Rating types 

> we can notice that The larger the risk, the higher the APR,

```python
plt.figure(figsize = [15,5])
order = ['AA', 'A', 'B', 'C', 'D', 'E', 'HR']
sb.violinplot(data = df2 , x = 'ProsperRating (Alpha)' , y ='BorrowerAPR' , order = order , color =color_base , inner = None);
plt.title('ProsperRating (Alpha) Types vs. Borrower APR ');
```

## Borrower APR based on Loan Original Amount 

> Obviously, the bigger the loan amount, the lower the APR.

```python
plt.figure(figsize=(15,5))
sb.regplot(data = df2 , y = 'BorrowerAPR' , x = 'LoanOriginalAmount' ,scatter_kws={'alpha': 1/10});
plt.title('Loan Original Amount Distribution vs. Borrower APR ');
```

## For all Prosper Rating types, the relationship between borrower APR and Duration 

> As the term gets bigger, the APR rises in the low risk category, whereas in the high risk rating , the APR falls.

```python
plt.figure(figsize=[9,5])
g = sb.pointplot(data = df2 , x ='ProsperRating (Alpha)' , y = 'BorrowerAPR' , hue = 'Term' , dodge = 0.4 , palette= 'Greens' ,linestyles="" , ci = 'sd' , order = order)
plt.title('ProsperRating (Alpha) vs Borrower APR Distribution in Each Term');
```

## the relationship between the APR and the Debt-to-Income Ratio (in Income Range= 1 to 25K  

> those with low incomes ranging from 1 to 25,000 chose larger debt with higher risks, resulting in increased debt-to-income ratio, which is why the borrower APR peaks at 0.36 .

```python
sub_data = df2.query('IncomeRange == "$1-24,999"')
g = sb.FacetGrid(data = sub_data , col='ProsperRating (Alpha)' , col_order = order   , margin_titles=True , col_wrap=3)
g.map(plt.scatter, 'BorrowerAPR', 'DebtToIncomeRatio');
plt.subplots_adjust(top=0.90);
g.fig.suptitle('The relationship between the APR Vs. the Debt-to-Income Ratiowhile (Income Range= 1 to 25K)');
```

```python
plt.figure(figsize=(8,5));
sub_data = df2[df2['ProsperRating (Alpha)'] =='HR'].query('IncomeRange == "$1-24,999"')
sb.regplot( data =sub_data , x= 'BorrowerAPR', y= 'DebtToIncomeRatio' ,fit_reg=False );
plt.title('The relationship between the APR Vs. the Debt-to-Income Ratiowhile (Income Range= 1 to 25K &  ProsperRating (Alpha) = HR)');
```
